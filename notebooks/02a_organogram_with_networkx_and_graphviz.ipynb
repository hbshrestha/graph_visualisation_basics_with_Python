# Using Networkx

````python
import networkx as nx

G = nx.DiGraph()

nodes = ["A","B","C","D","E","F","G","H"]

G.add_nodes_from(nodes)

G.add_edges_from([("A","B"), ("A","C"),
                  ("B","D"), ("B", "E"),
                  ("C", "F"), ("C", "G"), ("C","H")])

pos = {"A":(10, 10),
      "B":(7.5, 7.5), "C":(12.5, 7.5),
      "D":(6, 6), "E":(9, 6),
      "F":(11, 6), "G":(14, 6), "H":(17, 6)}

nx.draw_networkx(G, pos = pos, arrows = True,
                 node_shape = "s", node_color = "white")

````

````python
G = nx.DiGraph()

nodes = np.arange(0, 8).tolist()

G.add_nodes_from(nodes)

G.add_edges_from([(0,1), (0,2),
                  (1,3), (1, 4),
                  (2, 5), (2, 6), (2,7)])

pos = {0:(10, 10),
      1:(7.5, 7.5), 2:(12.5, 7.5),
      3:(6, 6), 4:(9, 6),
      5:(11, 6), 6:(14, 6), 7:(17, 6)}

labels = {0:"CEO",
         1:"Team A Lead", 2: "Team B Lead",
         3: "Staff A", 4: "Staff B",
         5: "Staff C", 6: "Staff D", 7: "Staff E"}

nx.draw_networkx(G, pos = pos, labels = labels, arrows = True,
                 node_shape = "s", node_color = "white")

plt.title("Organogram of Company X")
plt.savefig("../output/plain organogram using networkx.jpeg",
           dpi = 300)
plt.show()
````

<b>Degree centrality</b> of a node is the fraction of total nodes it is connected to. A node with high degree centrality is generally considered a highly active. For example, in a flight network, an airport with highest degree centrality would be the busiest airport. 
In `G`, node 3 i.e. Team B Lead has the highest degree centrality since it is connected to four other nodes.

<b>Betweenness centrality</b>: Betweenness centrality is a measure of how many times a particular node lies on a shortest path between two other nodes. In a flight network, an airport with highest betweenness centrality would be suitable to be a layover/connecting airport, because it lies mostly on the shortest route between two connected airports.

In `G`, team B lead has the highest betweenness centrality followed by team A lead. Other employees have 0 betweenness centrality since they lie on start or end of the edges.

<b>Closeness centrality</b>:
Closeness centrality is a measure of proximity of a node to other nodes. In flight network, the airports with high closeness centralities would most likely have several quick small flights outbound and inbound. In `G`, the staffs in bottom level have highest closeness centrality and are therefore, the more accessible/approachable ones. 

A node with high degree centrality will likely have higher betweenness centrality and closeness centrality as is the case with team leads in this example. 
It is to be noted that the betweenness centrality and closeness centrality values would change if `G` was an undirected graph while the degree centrality would remain the same.

## Getting the list of all the leafs/branches of the tree

````python
df = pd.DataFrame(nx.algorithms.all_simple_paths(G, start_node[0], end_nodes),
            columns = ["Root node", "Intermediate node", "Leaf node"])
df
````

````python
for index in df.index:
    for column in df.columns:
        value = df.loc[index, column]
        df.loc[index, column] = labels[value]        
df
````

````python
import graphviz

f = graphviz.Digraph(filename = "../output/plain organogram 1.gv")

names = ["A","B","C","D","E","F","G","H"]

positions = ["CEO","Team A Lead","Team B Lead", "Staff A","Staff B", "Staff C", "Staff D", "Staff E"]

for name, position in zip(names, positions):
    f.node(name, position)

    
#Specify edges
f.edge("A","B"); f.edge("A","C")   #CEO to Team Leads
f.edge("B","D"); f.edge("B","E")   #Team A relationship
f.edge("C","F"); f.edge("C","G"); f.edge("C","H")   #Team B relationship
    
f
````

````python
import graphviz

f = graphviz.Digraph(filename = "../output/plain_organogram_2.gv")

names = ["A","B","C","D","E","F","G","H"]

positions = ["CEO","Team A Lead","Team B Lead", "Staff A","Staff B", "Staff C", "Staff D", "Staff E"]

for name, position in zip(names, positions):
    if name == "A":
        f.node(name, position, shape = "oval")
        
    elif name in ["B","C"]:
        f.node(name, position, shape = "box")
    else:
        f.node(name, position, shape = "plaintext")

    
#Specify edges
f.edge("A","B"); f.edge("A","C")   #CEO to Team Leads
f.edge("B","D"); f.edge("B","E")   #Team A relationship
f.edge("C","F"); f.edge("C","G"); f.edge("C","H")   #Team B relationship
    
f
````
